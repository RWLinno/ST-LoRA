```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Load the data from the Excel file
file_path = 'param_study_new.xlsx'
df = pd.read_excel(file_path, engine='openpyxl', sheet_name='Sheet0')
print(df)
```

```python
cs={
    'b1': '#BADDF5',
    'y1': '#FFD99B',
    'g1': '#B4E2D3',
    'r1': '#FEA4AC',
    'r2': '#FFC3C7',
    'b2': '#5290E2',
    'b3': '#6e8fb2',
    'g2': '#7da494',
    'y2': '#eab67a',
    'r3': '#e5a79a',
    'r4': '#c16e71',
    'b4': '#abc8e5',
    'p1': '#d8a0c1',
    'p2': '#9f8db8',
    'g3': '#d0d08a'
}
```

```python
df1 = df.iloc[0:3, :].set_index(df.iloc[0:3, 0]).drop(df.columns[0], axis=1)
df2 = df.iloc[3:9, :].set_index(df.iloc[3:9, 0]).drop(df.columns[0], axis=1)
df3 = df.iloc[9:15, :].set_index(df.iloc[9:15, 0]).drop(df.columns[0], axis=1)
df1 = df1*100
df2 = df2*100
df3 = df3*100
print(df2)
```

```python
fig, axes = plt.subplots(1,3, figsize=(10, 4))
#plt.style.use('bmh')

df1.index = [5,20,100]
df1.iloc[:, 0].plot(ax=axes[0], marker='o', color=cs['g2'], linestyle=':')
df1.iloc[:, 1].plot(ax=axes[0], marker='^', color=cs['r3'], linestyle='--')
df1.iloc[:, 2].plot(ax=axes[0], marker='s', color=cs['b2'], linestyle='-.')
axes[0].set_xticks(df1.index)
axes[0].set_xticklabels(["5%", "10%", "100%"])
axes[0].set_xlabel('(a) Proportions of Dataset')
axes[0].set_ylabel('Metric Drop (%)')
axes[0].legend(loc='upper right')
axes[0].set_facecolor('#ffffe0')

df2.iloc[1:, 0].plot(ax=axes[1], marker='o', color=cs['g2'], linestyle=':')
df2.iloc[1:, 1].plot(ax=axes[1], marker='^', color=cs['r3'], linestyle='--')
df2.iloc[1:, 2].plot(ax=axes[1], marker='s', color=cs['b2'], linestyle='-.')
axes[1].set_xticks(df2.index[1:])
axes[1].set_xlabel('(b)Number of NALLs')
axes[1].set_ylabel('Metric Drop (%)')
axes[1].legend(loc='upper right')
axes[1].set_facecolor('#ffffe0')

df3.iloc[1:, 0].plot(ax=axes[2], marker='o', color=cs['g2'], linestyle=':')
df3.iloc[1:, 1].plot(ax=axes[2], marker='^', color=cs['r3'], linestyle='--')
df3.iloc[1:, 2].plot(ax=axes[2], marker='s', color=cs['b2'], linestyle='-.')
axes[2].set_xticks(df3.index[1:])
axes[2].set_xlabel('(c)Size of Intrinsic Rank')
axes[2].set_ylabel('Metric Drop (%)')
axes[2].legend(loc='upper right')
axes[2].set_facecolor('#ffffe0')

axes[0].grid(True)
axes[1].grid(True)
axes[2].grid(True)
plt.tight_layout()
# plt.show()

plt.savefig('./output/para_study_all.pdf', dpi=300, bbox_inches='tight')
```
